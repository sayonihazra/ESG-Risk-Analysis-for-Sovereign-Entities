```python
country=df.groupby('ESG Rating Rated Entity')['ESG Rating Overall (Num)'].max()
country.head(10)
```

```python
# Outliers using IQR method
q1 = df['ESG Rating Overall (Num)'].quantile(0.25)
q3 = df['ESG Rating Overall (Num)'].quantile(0.75)
iqr = q3 - q1

# Define outlier thresholds
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Filter outliers
outliers = df[(df['ESG Rating Overall (Num)'] < lower_bound) | 
              (df['ESG Rating Overall (Num)'] > upper_bound)]

# Print results
print("Outliers in ESG Rating Overall (Num):")
print(outliers['ESG Rating Overall (Num)'])
```

```python
#maps
#ESG laggard countries vs rating
sns.barplot( x='ESG Rating Overall (Num)',y='ESG Rating Rated Entity',data=bottom_10)
plt.title('ESG laggard countries vs rating')
plt.tight_layout()
plt.show()
```

```python
#ESG leader countries vs rating
sns.barplot( x='ESG Rating Overall (Num)',y='ESG Rating Rated Entity',data=top_10)
plt.title('ESG leader countries vs rating')
plt.tight_layout()
plt.show()
```
